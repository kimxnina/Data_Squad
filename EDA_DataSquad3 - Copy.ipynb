# Data Overview

```python
df = elective_df
# Create a new column 'timestamp' that combines 'lective' and 'section_term'
df['timestamp'] = df['lective'].astype(str) + '-' + df['section_term'].astype(str)
df.head()
```

```python
#Split Data into Semester and Trimester courses
semester_courses = df[df['section_term'].isin(['S1', 'S2'])]
trimester_courses = df[df['section_term'].isin(['T1', 'T2', 'T3', 'T4'])]
semester_courses = semester_courses.sort_values(by=['course_name', 'lective', 'section_term'], 
                                                 ascending=[True, True, True])
trimester_courses = trimester_courses.sort_values(by=['course_name', 'lective', 'section_term'], 
                                                 ascending=[True, True, True])
```

# Semester-Courses: Demand Bidding Primary List Group 1 + Group 2 per course

```python
# check if I can merge S1 and S2 bidding columns
mask_s1g1_s2g1 = ((semester_courses['demand_bidding_s1_g1_primary_list'] != 0) & (semester_courses['demand_bidding_s2_g1_primary_list'] == 0)) | ((semester_courses['demand_bidding_s1_g1_primary_list'] == 0) & (semester_courses['demand_bidding_s2_g1_primary_list'] != 0)) | ((semester_courses['demand_bidding_s1_g1_primary_list'] == 0) & (semester_courses['demand_bidding_s2_g1_primary_list'] == 0))
mask_s1g2_s2g2 = ((semester_courses['demand_bidding_s1_g2_primary_list'] != 0) & (semester_courses['demand_bidding_s2_g2_primary_list'] == 0)) | ((semester_courses['demand_bidding_s1_g2_primary_list'] == 0) & (semester_courses['demand_bidding_s2_g2_primary_list'] != 0)) | ((semester_courses['demand_bidding_s1_g2_primary_list'] == 0) & (semester_courses['demand_bidding_s2_g2_primary_list'] == 0))

combined_mask = mask_s1g1_s2g1 & mask_s1g2_s2g2

# Check if all rows meet the conditions
all_conditions_met = combined_mask.all()

print(f"All rows meet the condition: {all_conditions_met}")

# Print rows not meeting conditions with specified columns
cols_to_print = ['lective', 'section_term', 'course', 'course_name', 'demand_bidding_s1_g1_primary_list', 'demand_bidding_s1_g2_primary_list', 'demand_bidding_s2_g1_primary_list', 'demand_bidding_s2_g2_primary_list']
```

```python
#create a new column 'g1' which has the non-zero value from either 's1g1' or 's2g1' (or zero if both are zero), and a new column 'g2' which has the non-zero value from either 's1g2' or 's2g2' (or zero if both are zero)., then sort the values
semester_courses.loc[:, 'demand_bidding_g1_primary_list'] = semester_courses[['demand_bidding_s1_g1_primary_list', 'demand_bidding_s2_g1_primary_list']].max(axis=1)
semester_courses.loc[:, 'demand_bidding_g2_primary_list'] = semester_courses[['demand_bidding_s1_g2_primary_list', 'demand_bidding_s2_g2_primary_list']].max(axis=1)
```

```python
import matplotlib.pyplot as plt
import numpy as np
from matplotlib.colors import ListedColormap
import seaborn as sns

# Get a list of unique courses
courses = semester_courses['course_name'].unique()

# Define the pastel colormap
cmap = ListedColormap(['#ffd1dc', '#c2c2f0'])

# Calculate the number of rows required for the subplots
rows = len(courses) // 2 if len(courses) % 2 == 0 else len(courses) // 2 + 1

# Set up the subplot grid layout (2 columns) and adjust the figure size to make graphs smaller
fig, axs = plt.subplots(rows, 2, figsize=(12, len(courses)*4))

# Increase the vertical spacing
plt.subplots_adjust(hspace=0.5)

# Loop over all courses
for i, course in enumerate(courses):
    # Get the data for the current course
    course_data = semester_courses[semester_courses['course_name'] == course]
    
    # Group the data by timestamp and calculate the sum of 'demand_bidding_g1_primary_list' and 'demand_bidding_g2_primary_list'
    grouped_data = course_data.groupby('timestamp').agg({'demand_bidding_g1_primary_list': 'sum', 'demand_bidding_g2_primary_list': 'sum'}).reset_index()

    # Define the width of the bars
    bar_width = 0.35

    # Set the position of the bars on the x-axis
    r1 = np.arange(len(grouped_data['timestamp']))
    
    # Set the current subplot
    ax = axs[i // 2, i % 2] if len(courses) > 2 else axs[i % 2]

    # Create bars for 'demand_bidding_g1_primary_list' and 'demand_bidding_g2_primary_list'
    ax.bar(r1, grouped_data['demand_bidding_g1_primary_list'].values, color=cmap(0), width=bar_width, edgecolor='grey', label='demand_bidding_g1_primary_list')
    ax.bar(r1, grouped_data['demand_bidding_g2_primary_list'].values, bottom=grouped_data['demand_bidding_g1_primary_list'].values, color=cmap(1), width=bar_width, edgecolor='grey', label='demand_bidding_g2_primary_list')

    # Add a line representing the maximum number of students
    max_students = course_data['max_section_students'].max()
    ax.axhline(y=max_students, color='r', linestyle='-', label='Max students')

    # Add some text for labels, title and custom x-axis tick labels with increased fontsize
    ax.set_xlabel('Timestamp', fontsize=13)
    ax.set_ylabel('Demand Bidding', fontsize=13)
    ax.set_title(course, fontsize=12)
    ax.set_xticks(r1)
    ax.set_xticklabels(grouped_data['timestamp'], rotation=45, fontsize=12)
    ax.legend(fontsize=12)

# If the number of courses is odd, remove the last (empty) subplot
if len(courses) % 2:
    fig.delaxes(axs[-1, -1] if len(courses) > 2 else axs[-1])

# Adjust the layout so that the graphs don't overlap and display the graphs
plt.tight_layout()
plt.show()
```

## Heatmap for bidding demand

```python
# Group the data by course name and timestamp and calculate the sum of 'demand_bidding_g1_primary_list' and 'demand_bidding_g2_primary_list'
grouped_data = semester_courses.groupby(['course_name', 'timestamp']).agg({'demand_bidding_g1_primary_list': 'sum', 'demand_bidding_g2_primary_list': 'sum'}).reset_index()

# Add the two columns to get the total demand
grouped_data['total_demand'] = grouped_data['demand_bidding_g1_primary_list'] + grouped_data['demand_bidding_g2_primary_list']

# Pivot the data to get a matrix where each cell represents the total demand for a course at a specific timestamp
pivot_data = grouped_data.pivot(index='course_name', columns='timestamp', values='total_demand')

# Fill NaN values with 0
pivot_data = pivot_data.fillna(0)

# Draw the heatmap
plt.figure(figsize=(10, 30))  # Adjust the figure size as necessary
sns.heatmap(pivot_data, cmap="YlGnBu")

# Add a title to the heatmap
plt.title("Total Demand for Each Course at Each Timestamp")

# Show the plot
plt.show()
```
